```python
#importing countvectorizer from sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## Creating CountVectorizer with specified parameters
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
#This extracts the 'tag' column from the DataFrame (new_data) and converts it to a NumPy array of Unicode strings ('U'). This is necessary because CountVectorizer expects an array-like or iterable of strings.
```

```python
vector=cv.fit_transform(new_data['tag'].values.astype('U')).toarray()
```

```python
#shape of the vector
```

```python
vector.shape
```

cosine_similarity is a function from scikit-learn's metrics.pairwise module. It computes the cosine similarity between pairs of vectors.

```python
#importing cosine similarity from sklearn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#  'vector' is the array obtained from CountVectorizer
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
distance = sorted(list(enumerate(similarity[4])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
#Sorts the enumerated list of cosine similarity values in descending order based on the similarity values.
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])

    #Retrieves the top 5 most similar items along with their indices and print the list
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```

```python
import pandas as pd
from difflib import get_close_matches

# 'new_data' and 'similarity' are already defined

def recommend(movie_title, num_recommendations=5):
    # Find the closest match to the user input
    close_matches = get_close_matches(movie_title, new_data['title'], n=1, cutoff=0.8)

    if close_matches:
        closest_match = close_matches[0]
        print(f"Did you mean: {closest_match}?")

        # Get recommendations for the closest match
        index = new_data[new_data['title'] == closest_match].index[0]
        distance = sorted(enumerate(similarity[index]), reverse=True, key=lambda vector: vector[1])

        print(f"Top {num_recommendations} recommendations for {closest_match}:")
        for i in distance[1:num_recommendations+1]:  # Start from 1 to exclude the input movie
            recommended_movie = new_data.iloc[i[0]].title
            print(recommended_movie)
    else:
        print(f"No close match found for '{movie_title}' in the dataset.")

# Get user input
user_input = input("Enter a movie title: ")

# Call the recommend function with user input
recommend(user_input)
```
